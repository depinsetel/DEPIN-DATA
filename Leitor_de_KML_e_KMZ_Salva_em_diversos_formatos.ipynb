```python
import zipfile
import geopandas as gpd

# Extrair o arquivo KML do arquivo KMZ
with zipfile.ZipFile("Infovias Estaduais.kmz", "r") as kmz:
    kml_files = [name for name in kmz.namelist() if name.endswith('.kml')]
    if len(kml_files) != 1:
        raise ValueError("O arquivo KMZ deve conter exatamente um arquivo KML.")
    kml_file = kml_files[0]
    kmz.extract(kml_file, path=".")

# Ler o arquivo KML com o GeoPandas
my_map = gpd.read_file(kml_file)

# Exibir o GeoDataFrame
display(my_map)

```

```python
# Salvar o GeoDataFrame como um arquivo Shapefile
my_map.to_file("my_map.shp")

# Salvar o GeoDataFrame como um arquivo GeoJSON
my_map.to_file("my_map.geojson", driver='GeoJSON')

# Salvar o GeoDataFrame como um arquivo GeoPackage (.gpkg)
my_map.to_file("my_map.gpkg", driver='GPKG')

# Salvar o GeoDataFrame como um arquivo Esri File Geodatabase (.gdb)
# my_map.to_file("my_map.gdb", driver='FileGDB')

# Salvar o GeoDataFrame como um arquivo CSV (Comma-Separated Values)
my_map.to_csv("my_map.csv")

# Salvar o GeoDataFrame como um arquivo Excel (.xlsx)
my_map.to_excel("my_map.xlsx")
```
